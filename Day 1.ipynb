```python
import pandas as pd
data = pd.read_csv("input.txt", sep="   ", header=None)
data.columns = ["List One", "List Two"]
data
```

```python
print(data["List One"].min())
print(data["List One"].max())
print(data["List Two"].min())
print(data["List Two"].max())
```

```python
data.sort_values(by=["List One"], ascending=True, inplace=True)
data
```

```python
list_one = data["List One"].tolist()
```

```python
print(list_one)
```

```python
data.sort_values(by=["List Two"], ascending=True, inplace=True)
list_two = data["List Two"].tolist()
```

```python
print(list_two)
```

```python
print(len(list_one))
print(len(list_two))
```

```python
import numpy as np

array1 = np.array(list_one)
array2 = np.array(list_two)

result = abs(array2 - array1)
print(result.sum())
```

```python
duplicates = []

for val in array1:
    if val in array2:
        duplicates.append(val)
```

```python
duplicates
```

```python
from collections import Counter

c = Counter(array2)
listC = [c.get(i) for i in duplicates]
print(listC)
```

```python
array3 = np.array(listC)
result = (duplicates * array3)
print(result.sum())
```
