```python
import pandas as pd
data = pd.read_csv("input2.txt", header=None)
data.columns = ["Reports"]
data["Reports"] = data["Reports"].str.split()
data
```

```python
reports = data["Reports"].tolist()
reports
```

```python
reports = [[int(int(j)) for j in i] for i in reports]
reports
```

```python
increasing = []

for report in reports:
    res = all(i < j for i, j in zip(report, report[1:]))
    if True:
        increasing.append(res)
```

```python
decreasing = []

for report in reports:
    res = all(i > j for i, j in zip(report, report[1:]))
    if True:
        decreasing.append(res)
```

```python
df = pd.DataFrame()
df['Reports'] = pd.Series(reports)
df['Increasing'] = pd.Series(increasing)
df['Decreasing'] = pd.Series(decreasing)
df
```

```python
reports = df["Reports"].tolist()
```

```python
difference = []

for report in reports:
    res = [j-i for i, j in zip(report, report[1:])]
    difference.append(res)
```

```python
diff_abs = []

for report in difference:
    res = [abs(ele) for (ele) in report]
    diff_abs.append(res)
```

```python
diff_abs
```

```python
len(diff_abs)
```

```python
flags = [(min(i) >= 1 and max(i) <= 3) for i in diff_abs]
```

```python
flags
```

```python
df['Flags'] = pd.Series(flags)
df
```

```python
import numpy as np

df["IncreaseFlag"] = np.logical_or(df["Increasing"], df["Flags"])
df["DecreaseFlag"] = np.logical_or(df["Decreasing"], df["Flags"])
df["ProblemDampener"] = np.logical_or(df["IncreaseFlag"], df["DecreaseFlag"])
```

```python
df = df[df["ProblemDampener"] == True]
df
```
